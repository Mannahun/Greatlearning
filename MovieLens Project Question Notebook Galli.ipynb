### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
#your code here
grouped_item= item.groupby('year').sum()
grouped_item.tail()
```

```python
grouped_item['total releases']= grouped_item[['Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']].sum(axis=1)
grouped_item.tail()
```

```python
grouped_item.drop('movie id', axis=1, inplace=True)
grouped_item.tail()
```

```python
grouped_item.drop('unknown', axis=1, inplace=True)
grouped_item.tail()
```

```python
grouped_item.tail()
```

```python
percent= grouped_item.apply(lambda x: x/grouped_item['total releases']*100)
```

```python
percent.tail()
```

```python
percent.drop('total releases', axis=1, inplace=True)
percent.tail()
```

```python
sns.heatmap(percent)
```

**Insights:**
Some genres, such as Comedy and Drama, have been highly represented over the years. Drama appeared to peak in the early age of cinema and then again around 1968, while, interestingly, horror saw an increase around 1976. A few genres, such as fantasy or documentary always made up a very small percentage of releases, according to the available data. A genre like Western was well represented in the '60s and pretty much disappeared in more recent years.

```python
#your code here
movie= data.groupby('movie id').count()
movie.head()
```

```python
avg_movie= data.groupby('movie id').mean()
avg_movie.columns=['user id', 'avg_rating', 'timestamp']
avg_movie.head()
```

```python
movie['avg_rating']= avg_movie['avg_rating']
movie.head()
```

```python
movie=movie.sort_values(by='rating', ascending=False)
movie.head()
```

```python
best_movies= movie.loc[movie['rating']>100].sort_values(by='avg_rating', ascending=False)
best_movies=best_movies.head(25)
```

```python
best_movies.shape
```

```python
new_item= item.set_index('movie id')
new_item.head()
```

```python
solution=best_movies.merge(new_item, right_index=True, left_index=True, how='inner')
solution.drop(['Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'], axis=1)
```
