# Preparing the data for the models

## Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
df['age']= scaler.fit_transform(df[['age']])
```

```python
df.balance= scaler.fit_transform(df[['balance']])
df.duration=scaler.fit_transform(df[['duration']])
df.previous=scaler.fit_transform(df[['previous']])
```

## One hot encoding of categorical variables

```python
df= pd.get_dummies(df, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'pdays'], drop_first=True)
df.head()
```

```python
df= pd.get_dummies(df, columns=['Target'], drop_first=True)
df.head(1)
```

```python
df.shape
```

## Creating datasets for the models

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Target_yes', axis=1)
y=df.Target_yes
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Logistic Regression

```python
from sklearn import metrics

from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
coef_df = pd.DataFrame(model.coef_)
coef_df['intercept'] = model.intercept_
print(coef_df)
```

```python
print("Training accuracy",model.score(X_train,y_train))  # this will give training accuracy as training data has been used
print()
print("Testing accuracy",model.score(X_test, y_test))   # this will give testing accuracy as testing data has been used
```

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score, roc_auc_score
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels = [0,1] , yticklabels = [0,1] )
plt.ylabel('Observed')
plt.xlabel('Predicted')
plt.show()
```

```python
acc_logr= model.score(X_test, y_test)
recall_logr= recall_score(y_test, y_predict)
prec_logr= precision_score(y_test, y_predict)
f1_logr=f1_score(y_test, y_predict)
```

```python
resultsDf = pd.DataFrame({'Method':['Logistic regression'], 'accuracy': acc_logr, 'recall': recall_logr, 'precision': prec_logr, 'F1': f1_logr})
resultsDf = resultsDf[['Method', 'accuracy', 'recall', 'precision', 'F1']]
resultsDf
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dTree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3, random_state=1)
dTree.fit(X_train, y_train)
print("Training accuracy :", dTree.score(X_train, y_train))
print("Testing accuracy :", dTree.score(X_test, y_test))
```

```python
print(dTree.score(X_test , y_test))
y_predict = dTree.predict(X_test)

cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels = [0,1] , yticklabels = [0,1] )
plt.ylabel('Observed')
plt.xlabel('Predicted')
plt.show()
```

```python
acc_dt= model.score(X_test, y_test)
recall_dt= recall_score(y_test, y_predict)
prec_dt= precision_score(y_test, y_predict)
f1_dt=f1_score(y_test, y_predict)
```

```python
tempResultsDf = pd.DataFrame({'Method':['Decision Tree'], 'accuracy': acc_dt, 'recall': recall_dt, 'precision': prec_dt, 'F1': f1_dt})
resultsDf = pd.concat([resultsDf, tempResultsDf])
resultsDf = resultsDf[['Method', 'accuracy', 'recall', 'precision', 'F1']]
resultsDf
```
